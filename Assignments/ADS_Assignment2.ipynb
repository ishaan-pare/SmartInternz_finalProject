```python
#pie chart
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = ["0-20", "20-40", "40-80", "80-100"]
x = [0,0,0,0]
for i in data["age"]:
    if i<=20:
        x[0]+=1
    elif i<=40:
        x[1]+=1
    elif i<=80:
        x[2]+=1
    else:
        x[3]+=1
```

```python
plt.pie(x, labels=labels, autopct="%1.1f%%")
```
